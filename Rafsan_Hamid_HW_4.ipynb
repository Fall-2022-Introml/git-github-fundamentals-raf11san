```python
#Please make sure you use the exact names for variables including the casing of letters
y2_pred = [1] * len(y2)

from sklearn import metrics

accu = metrics.accuracy_score(y2, y2_pred)
prec = metrics.precision_score(y2, y2_pred)
recall = metrics.recall_score(y2, y2_pred)
fscore = metrics.f1_score(y2, y2_pred)

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X2_st = scaler.fit_transform(X2)

print(fscore)
```
